#**Soal 2, 3 dan 4**



>menghitung BMI



```python
BMI = data.berat_badan_kg/(data.tinggi_badan_cm/100)**2
```

# **Soal 5**

> fungsi mencari status BMI

```python
def StatusBMI(BMI):
  if BMI >= 18.5 and BMI <= 24.9:
    return 'Normal'
  elif BMI >= 25.0 and BMI <= 29.9:
    return 'Overweight'
  elif BMI >= 30.0 and BMI <= 34.9:
    return 'Obesity'
  elif BMI < 18.5:
    return 'Underweight'
```

> membuat kolom Status BMI dan apply fungsi StatusBMI()

```python
data['Status_BMI'] = data.BMI.apply(StatusBMI)
```

> cek colom baru

```python
data.head()
```
